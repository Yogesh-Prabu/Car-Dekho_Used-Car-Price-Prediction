```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load your dataset (update the path accordingly)
df_scaled = pd.read_csv(r'SBS_Processed_Datasets/04_1_scaled_dataframe.csv')
```

```python
# Visualization column mappings
visualization_columns = {
    "price": "Price",
    "kms_driven": "Kilometers Driven",
    "model_year": "Model Year",
    "mileage_kmpl": "Mileage (kmpl)",
    "engine_cc": "Engine Capacity (cc)",
    "car_age": "Car Age"
}

# Distribution Plots for Key Columns
for col, label in visualization_columns.items():
    plt.figure(figsize=(10, 6))
    sns.histplot(df_scaled[col], kde=True, bins=30)
    plt.title(f"Distribution of {label}")
    plt.xlabel(label)
    plt.ylabel("Frequency")
    plt.show()

# Select Numeric Columns
df_numeric = df_scaled.select_dtypes(include=[np.number])
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Scaled Features")
plt.show()
```

```python
# Scatter Plots: Price vs Key Features
scatter_features = ["kms_driven", "model_year", "mileage_kmpl", "engine_cc", "car_age"]

for feature in scatter_features:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=feature, y="price", data=df_scaled)
    plt.title(f"Price vs {visualization_columns[feature]}")
    plt.xlabel(visualization_columns[feature])
    plt.ylabel("Price")
    plt.show()
```
